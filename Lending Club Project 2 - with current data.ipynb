# Feature Cleaning and Engineering

## Categorize Features

```python
train_df.get_dtype_counts()
```

```python
test_df.get_dtype_counts()
```

```python
cols = df.columns.tolist()
all_null_feature = []
num_feature = []
ob_feature = []

for col in cols:
    if df[col].isnull().sum() == df.shape[0]:
        all_null_feature.append(col)
    
    else:
        if df[col].dtype == 'object':
            ob_feature.append(col)
        else:
            num_feature.append(col)
```

## Remove identical features

```python
df[['loan_amnt', 'funded_amnt','from_which_dataset']]
```

We can see that Loan amount and funded amount are the same in Loan2015 data, so we will remove 
one feature as it will not be helpful in training our model.

```python
train_df = train_df.drop(columns = 'funded_amnt')
```

```python
test_df = test_df.drop(columns = 'funded_amnt')
```

```python
print("The train_df has %d rows and %s columns." % (train_df.shape[0], train_df.shape[1]))
print("The test_df has %d rows and %s columns." % (test_df.shape[0], test_df.shape[1]))
```

### Keep rows that 'term' = 36 month & 'loan_status' either equal to fully paid or charged off.

* term = 36 month, is because for now we know the result of the loan.
* Loan_status = fully paid or charged off, is because we make it a binary classfication problem. And, we would like 
                                            to choose loans that will be fully paied.

```python
train_df = train_df[(train_df['loan_status'] == 'Fully Paid') |
   (train_df['loan_status'] == 'Charged Off') ]

train_df['term'].replace(' 36 months', int(36), inplace=True)
train_df = train_df.query("term == 36")
train_df = train_df.drop(columns = 'term')
```

```python
ob_feature.remove('term')
```

```python
print("The train_df has %d rows and %s columns." % (train_df.shape[0], train_df.shape[1]))
print("The test_df has %d rows and %s columns." % (test_df.shape[0], test_df.shape[1]))
```

```python
train_df['loan_status'] = train_df.loan_status.map({'Fully Paid':0, 'Charged Off':1})

train_df.loan_status.value_counts(normalize=1)
```

```python
ob_feature.remove('loan_status')
num_feature.append('loan_status')
```

### Date to numerical number: ‘earliest_cr_line’

```python
train_df['earliest_cr_line'].head(5)
```

```python
train_cr_month = train_df.earliest_cr_line.apply(lambda x: x.split('-')[0])
train_cr_year = train_df.earliest_cr_line.apply(lambda x: x.split('-')[1])

train_dic_month = {'Jan':11, 'Feb':10, 'Mar':9, 'Apr':8,'May':7, 'Jun':6, 'Jul':5, 'Aug':4, 'Sep':3, 'Oct':2, 'Nov':1, 'Dec':0}
train_df['earliestcrline_month'] = train_df.earliest_cr_line.apply(lambda x: (2015 - int(x.split('-')[1]))*12 + train_dic_month[x.split('-')[0]])

train_df = train_df.drop(columns = 'earliest_cr_line')
```

```python
test_df['earliestcrline_month'] = test_df.earliest_cr_line.apply(lambda x: 
                                        (now.year - int(x.split('-')[0])) * 12 + 
                                        (now.month - int(x.split('-')[1])))                                                        
```

### Model Tuning

```python
import gc
import copy
gc.collect()

print('Starting......')

default_params = {}

default_params['eta'] = 0.08
default_params['max_depth'] = 5
default_params['min_child_weight'] = 2
default_params['n_estimator'] = 1000
default_params['gamma'] = 0.1
default_params['subsample'] = 0.8
default_params['colsample_bytree'] = 0.8
default_params['objective'] = 'binary:logistic'
default_params['scale_pos_weight'] = 1

params_test = {}

# params_test['eta'] = 0.08
# params_test['max_depth'] = 5
# params_test['min_child_weight'] = 2 
params_test['n_estimator'] = range(800,2000,300)
# params_test['gamma'] = 0.1
# params_test['subsample'] = 0.8
params_test['colsample_bytree'] = np.arange(0.2, 1, 0.2)

greater_is_better = True
best_xgb_params = copy.copy(default_params)

count = 0
for p in params_test:
    
    num = count + 1
    
    print(str(num), ': Tuning parameter %s' % p)
    
    params = best_xgb_params
    scores = []
    
    xgb = XGBClassifier(**params)
    
    to_test = {p:params_test[p]}
    GS = GridSearchCV(estimator=xgb, param_grid = to_test, cv=4, scoring='roc_auc')
    GS.fit(X_train, y_train)
    print(' ----> Best Score is %s' % (GS.best_score_))
    print(' ----> Best parameter for %s is %s' % (p, GS.best_params_))
    best_xgb_params[p] = GS.best_params_[p]
```

## Roc-auc curve

```python
from sklearn.metrics import roc_curve, auc
from sklearn import linear_model, datasets
import pylab as pl
def draw_ROC(model, dtrain, dvalid, dtest, y_train, y_valid, y_test ):
    probas_ = model.predict(dvalid, ntree_limit=model.best_ntree_limit)
    probas_1 = model.predict(dtrain, ntree_limit=model.best_ntree_limit)
    probas_2 = model.predict(dtest, ntree_limit=model.best_ntree_limit)
    fpr, tpr, thresholds = roc_curve(y_valid, probas_)
    fpr_1, tpr_1, thresholds_1 = roc_curve(y_train, probas_1)
    fpr_2, tpr_2, thresholds_2 = roc_curve(y_test, probas_2)
    roc_auc = auc(fpr, tpr)
    roc_auc_1 = auc(fpr_1, tpr_1)
    roc_auc_2 = auc(fpr_2, tpr_2)
    print("Area under the ROC curve - validation: %f" % roc_auc)
    print("Area under the ROC curve - train: %f" % roc_auc_1)
    print("Area under the ROC curve - test: %f" % roc_auc_2)
    # Plot ROC curve
    plt.figure(figsize=(8,8))
    plt.plot(fpr, tpr, label='ROC curve - valid(AUC = %0.2f)' % roc_auc, color='r')
    plt.plot(fpr_1, tpr_1, label='ROC curve - train (AUC = %0.2f)' % roc_auc_1, color='b')
    plt.plot(fpr_2, tpr_2, label='ROC curve - test (AUC = %0.2f)' % roc_auc_2, color='g')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC for lead score model')
    plt.legend(loc="lower right")
    plt.show()
```

```python
draw_ROC(gbm, dtrain, dvalid, dtest, y_train, y_valid, y_test)
```
